## Problem 1-5: Comparing Two Classes

Two classes have the same mean exam score of 80 points, but their distributions differ:

**Class A:** 77, 78, 79, 80, 81, 82, 83  
**Class B:** 60, 70, 75, 80, 85, 90, 100

Let's calculate all measures of variability!

````r
# Define the data
class_a <- c(77, 78, 79, 80, 81, 82, 83)
class_b <- c(60, 70, 75, 80, 85, 90, 100)

# Verify means
mean_a <- mean(class_a)
mean_b <- mean(class_b)

cat("Class A mean:", mean_a, "\n")
cat("Class B mean:", mean_b, "\n\n")

# Calculate range for both classes
range_a <- diff(range(class_a))  # max - min
range_b <- diff(range(class_b))

cat("RANGE:\n")
cat("Class A: min =", min(class_a), ", max =", max(class_a), ", range =", range_a, "\n")
cat("Class B: min =", min(class_b), ", max =", max(class_b), ", range =", range_b, "\n\n")

# Calculate variance and standard deviation
var_a <- var(class_a)
sd_a <- sd(class_a)
var_b <- var(class_b)
sd_b <- sd(class_b)

cat("VARIANCE AND STANDARD DEVIATION:\n")
cat("Class A: sÂ² =", round(var_a, 2), ", s =", round(sd_a, 2), "\n")
cat("Class B: sÂ² =", round(var_b, 2), ", s =", round(sd_b, 2), "\n\n")

cat("ðŸ’¡ Interpretation:\n")
cat("   Class A has", round(range_a), "points range and s =", round(sd_a, 2), "(low variability)\n")
cat("   Class B has", round(range_b), "points range and s =", round(sd_b, 2), "(high variability)\n")
cat("   Class A is MORE consistent despite having the same mean!\n")
````

````r
# Visualize the difference in variability

# Create data frames
df_a <- data.frame(score = class_a, class = "Class A")
df_b <- data.frame(score = class_b, class = "Class B")
df_combined <- rbind(df_a, df_b)

# Plot 1: Dot plots showing spread
p1 <- ggplot(df_combined, aes(x = score, y = class, color = class)) +
  geom_point(size = 4, alpha = 0.7) +
  geom_vline(xintercept = 80, linetype = "dashed", color = "black", linewidth = 0.8) +
  labs(title = "Comparison of Score Distributions",
       x = "Score",
       y = "") +
  annotate("text", x = 80, y = 2.5, label = "Mean = 80", size = 3) +
  scale_color_manual(values = c("Class A" = "steelblue", "Class B" = "coral")) +
  theme_minimal() +
  theme(legend.position = "none")

# Plot 2: Boxplots
p2 <- ggplot(df_combined, aes(x = class, y = score, fill = class)) +
  geom_boxplot(alpha = 0.7) +
  geom_hline(yintercept = 80, linetype = "dashed", color = "black", linewidth = 0.8) +
  labs(title = "Boxplots Showing Variability",
       x = "",
       y = "Score") +
  scale_fill_manual(values = c("Class A" = "steelblue", "Class B" = "coral")) +
  theme_minimal() +
  theme(legend.position = "none")

# Display plots
grid.arrange(p1, p2, ncol = 2)

cat("\nðŸ“Š Notice:\n")
cat("   â€¢ Class A scores cluster tightly around the mean (low spread)\n")
cat("   â€¢ Class B scores are spread out widely (high variability)\n")
cat("   â€¢ Both have the same mean, but very different consistency!\n")
````

## Problem 2-4: Calculating Deviations and Variance

Let's see how to calculate variance step-by-step for Class A.

````r
# Create a detailed table of calculations for Class A
class_a_table <- data.frame(
  xi = class_a,
  deviation = class_a - mean_a,
  deviation_squared = (class_a - mean_a)^2
)

cat("DEVIATION TABLE FOR CLASS A (mean = 80):\n")
cat("==========================================\n\n")
print(class_a_table, row.names = FALSE)

# Calculate sums
sum_deviations <- sum(class_a_table$deviation)
sum_squared_deviations <- sum(class_a_table$deviation_squared)

cat("\n==========================================\n")
cat("Sum of deviations:", round(sum_deviations, 10), "(always equals 0!)\n")
cat("Sum of squared deviations:", sum_squared_deviations, "\n\n")

# Calculate variance
n <- length(class_a)
variance_manual <- sum_squared_deviations / (n - 1)
sd_manual <- sqrt(variance_manual)

cat("Sample size: n =", n, "\n")
cat("Variance: sÂ² =", sum_squared_deviations, "/ (n-1) =", 
    sum_squared_deviations, "/", (n-1), "=", round(variance_manual, 2), "\n")
cat("Standard deviation: s = âˆšsÂ² = âˆš", round(variance_manual, 2), "=", 
    round(sd_manual, 2), "\n\n")

# Verify with R's built-in functions
cat("Verification using R functions:\n")
cat("var(class_a) =", var(class_a), "âœ“\n")
cat("sd(class_a) =", sd(class_a), "âœ“\n")
````

````r
# Now do the same for Class B
class_b_table <- data.frame(
  xi = class_b,
  deviation = class_b - mean_b,
  deviation_squared = (class_b - mean_b)^2
)

cat("DEVIATION TABLE FOR CLASS B (mean = 80):\n")
cat("==========================================\n\n")
print(class_b_table, row.names = FALSE)

sum_deviations_b <- sum(class_b_table$deviation)
sum_squared_deviations_b <- sum(class_b_table$deviation_squared)

cat("\n==========================================\n")
cat("Sum of deviations:", round(sum_deviations_b, 10), "\n")
cat("Sum of squared deviations:", sum_squared_deviations_b, "\n\n")

variance_b_manual <- sum_squared_deviations_b / (n - 1)
sd_b_manual <- sqrt(variance_b_manual)

cat("Variance: sÂ² =", round(variance_b_manual, 2), "\n")
cat("Standard deviation: s =", round(sd_b_manual, 2), "\n\n")

cat("ðŸ“ˆ Comparison:\n")
cat("   Class A: sÂ² =", round(var_a, 2), ", s =", round(sd_a, 2), "\n")
cat("   Class B: sÂ² =", round(var_b, 2), ", s =", round(sd_b, 2), "\n")
cat("   Class B has", round(var_b / var_a, 1), "times more variance!\n")
````

## Problem 6: Computational Formula

The computational formula is often easier for hand calculation:
$$s^2 = \frac{\sum x_i^2 - \frac{(\sum x_i)^2}{n}}{n-1}$$

````r
# Verify Class A variance using computational formula

sum_xi <- sum(class_a)
sum_xi_squared <- sum(class_a^2)
n <- length(class_a)

cat("COMPUTATIONAL FORMULA VERIFICATION\n")
cat("====================================\n\n")
cat("Given:\n")
cat("  Î£xi =", sum_xi, "\n")
cat("  Î£xiÂ² =", sum_xi_squared, "\n")
cat("  n =", n, "\n\n")

# Calculate using computational formula
variance_comp <- (sum_xi_squared - (sum_xi^2 / n)) / (n - 1)

cat("Using computational formula:\n")
cat("sÂ² = [", sum_xi_squared, "- (", sum_xi, ")Â² /", n, "] / (n-1)\n")
cat("sÂ² = [", sum_xi_squared, "-", round(sum_xi^2 / n, 2), "] /", (n-1), "\n")
cat("sÂ² =", round(variance_comp, 2), "\n\n")

cat("Comparison:\n")
cat("  Definitional formula: sÂ² =", round(var_a, 2), "\n")
cat("  Computational formula: sÂ² =", round(variance_comp, 2), "\n")
cat("  âœ“ Both methods give the same result!\n")
````

## Problem 7: Interquartile Range (IQR)

Test scores for 12 students (out of 40): 24, 26, 28, 30, 31, 32, 33, 34, 35, 36, 37, 40

````r
# Test scores data
test_scores <- c(24, 26, 28, 30, 31, 32, 33, 34, 35, 36, 37, 40)

# Calculate quartiles
q1 <- quantile(test_scores, 0.25)
q2 <- median(test_scores)
q3 <- quantile(test_scores, 0.75)
iqr_value <- IQR(test_scores)

# Also calculate range and standard deviation for comparison
range_value <- diff(range(test_scores))
sd_value <- sd(test_scores)

cat("TEST SCORES ANALYSIS\n")
cat("====================\n\n")
cat("Data:", test_scores, "\n\n")

cat("Five-Number Summary:\n")
cat("  Min  =", min(test_scores), "\n")
cat("  Q1   =", q1, "\n")
cat("  Q2   =", q2, "(median)\n")
cat("  Q3   =", q3, "\n")
cat("  Max  =", max(test_scores), "\n\n")

cat("Measures of Spread:\n")
cat("  Range = Max - Min =", max(test_scores), "-", min(test_scores), "=", range_value, "\n")
cat("  IQR   = Q3 - Q1   =", q3, "-", q1, "=", iqr_value, "\n")
cat("  SD    =", round(sd_value, 2), "\n\n")

cat("ðŸ’¡ Interpretation:\n")
cat("   â€¢ The middle 50% of scores span", iqr_value, "points (from", q1, "to", q3, ")\n")
cat("   â€¢ IQR is resistant to outliers, unlike range\n")
cat("   â€¢ IQR measures spread of central data\n")
````

````r
# Visualize with boxplot
scores_df <- data.frame(score = test_scores)

ggplot(scores_df, aes(y = score)) +
  geom_boxplot(fill = "lightgreen", color = "black", width = 0.5) +
  geom_hline(yintercept = q1, linetype = "dashed", color = "blue", alpha = 0.5) +
  geom_hline(yintercept = q2, linetype = "dashed", color = "red", alpha = 0.5) +
  geom_hline(yintercept = q3, linetype = "dashed", color = "blue", alpha = 0.5) +
  labs(title = "Test Scores Boxplot",
       y = "Score (out of 40)") +
  annotate("text", x = 0.35, y = q1, label = paste("Q1 =", q1), size = 3) +
  annotate("text", x = 0.35, y = q2, label = paste("Q2 =", q2), size = 3, color = "red") +
  annotate("text", x = 0.35, y = q3, label = paste("Q3 =", q3), size = 3) +
  annotate("text", x = -0.3, y = (q1 + q3)/2, 
           label = paste("IQR =", iqr_value), size = 3.5, fontface = "bold") +
  theme_minimal() +
  theme(axis.text.x = element_blank(),
        axis.ticks.x = element_blank())
````

## Problem 8: Interpreting Standard Deviation

Two stores track daily customer counts:

**Store A:** mean = 120 customers, s = 8  
**Store B:** mean = 120 customers, s = 35

````r
# Simulate data for visualization
set.seed(123)

# Generate data approximately matching the given statistics
store_a <- round(rnorm(30, mean = 120, sd = 8))
store_b <- round(rnorm(30, mean = 120, sd = 35))

# Adjust to ensure they have exactly mean = 120
store_a <- store_a - mean(store_a) + 120
store_b <- store_b - mean(store_b) + 120

cat("COMPARING TWO STORES\n")
cat("====================\n\n")

cat("Store A (Low Variability):\n")
cat("  Mean =", round(mean(store_a), 1), "\n")
cat("  SD   =", round(sd(store_a), 1), "\n")
cat("  Range: mean Â± 2s = [", round(120 - 2*8, 1), ",", round(120 + 2*8, 1), "]\n\n")

cat("Store B (High Variability):\n")
cat("  Mean =", round(mean(store_b), 1), "\n")
cat("  SD   =", round(sd(store_b), 1), "\n")
cat("  Range: mean Â± 2s = [", round(120 - 2*35, 1), ",", round(120 + 2*35, 1), "]\n\n")

cat("ðŸ’¡ Interpretation:\n")
cat("   â€¢ Store A: Consistent traffic (104-136 customers most days)\n")
cat("   â€¢ Store B: Highly variable traffic (50-190 customers)\n")
cat("   â€¢ Store A is easier to staff and manage\n")
````

## Problem 10: Range vs IQR with Outliers

Monthly expenses (in dollars): 500, 650, 700, 750, 800, 850, 900, 950, 1000, 8500

````r
# Monthly expenses with one extreme outlier
expenses <- c(500, 650, 700, 750, 800, 850, 900, 950, 1000, 8500)

# Calculate all measures
mean_exp <- mean(expenses)
median_exp <- median(expenses)
range_exp <- diff(range(expenses))
iqr_exp <- IQR(expenses)
sd_exp <- sd(expenses)

cat("MONTHLY EXPENSES WITH OUTLIER\n")
cat("==============================\n\n")
cat("Data:", expenses, "\n\n")

cat("Measures of Center:\n")
cat("  Mean   = $", round(mean_exp, 2), "\n")
cat("  Median = $", median_exp, "\n\n")

cat("Measures of Spread:\n")
cat("  Range = Max - Min = $", max(expenses), "- $", min(expenses), "= $", range_exp, "\n")
cat("  IQR   = Q3 - Q1   = $", quantile(expenses, 0.75), "- $", quantile(expenses, 0.25), 
    "= $", iqr_exp, "\n")
cat("  SD    = $", round(sd_exp, 2), "\n\n")

# Compare without outlier
expenses_no_outlier <- expenses[expenses < 8000]
range_no_outlier <- diff(range(expenses_no_outlier))
iqr_no_outlier <- IQR(expenses_no_outlier)

cat("WITHOUT the $8,500 outlier:\n")
cat("  Range = $", range_no_outlier, "\n")
cat("  IQR   = $", iqr_no_outlier, "\n\n")

cat("ðŸ“Š Impact of Outlier:\n")
cat("   â€¢ Range changed from $", range_no_outlier, "to $", range_exp, 
    "(", round((range_exp - range_no_outlier)/range_no_outlier * 100), "% increase!)\n")
cat("   â€¢ IQR changed from $", iqr_no_outlier, "to $", iqr_exp, 
    "(", round(abs(iqr_exp - iqr_no_outlier)), "difference)\n")
cat("   â€¢ IQR is RESISTANT to outliers, Range is SENSITIVE\n")
````

````r
# Visualize the impact of the outlier
expenses_df <- data.frame(
  amount = expenses,
  outlier = ifelse(expenses > 5000, "Outlier", "Normal")
)

p1 <- ggplot(expenses_df, aes(x = 1:length(expenses), y = amount, color = outlier)) +
  geom_point(size = 4) +
  geom_hline(yintercept = median_exp, linetype = "dashed", color = "blue") +
  labs(title = "Monthly Expenses with Extreme Outlier",
       x = "Month",
       y = "Expense ($)") +
  scale_color_manual(values = c("Normal" = "steelblue", "Outlier" = "red")) +
  annotate("text", x = 5, y = median_exp + 500, 
           label = paste("Median = $", median_exp), color = "blue") +
  theme_minimal() +
  theme(legend.position = "top")

p2 <- ggplot(expenses_df, aes(y = amount)) +
  geom_boxplot(fill = "lightblue", outlier.color = "red", outlier.size = 3) +
  labs(title = "Boxplot Highlights Outlier",
       y = "Expense ($)") +
  annotate("text", x = 0.3, y = 8500, label = "Outlier!", color = "red", fontface = "bold") +
  theme_minimal() +
  theme(axis.text.x = element_blank(),
        axis.ticks.x = element_blank())

grid.arrange(p1, p2, ncol = 2)

cat("\nðŸ’¡ Key Lesson:\n")
cat("   Use IQR when outliers are present - it's RESISTANT\n")
cat("   Range is useful but highly SENSITIVE to extreme values\n")
````

## Problems 11-12: Effect of Transformations

Study hours per week for 5 students: 15, 20, 25, 30, 35

What happens when we add/multiply by a constant?

````r
# Original data
hours <- c(15, 20, 25, 30, 35)

mean_orig <- mean(hours)
sd_orig <- sd(hours)
var_orig <- var(hours)

cat("ORIGINAL DATA:\n")
cat("Data:", hours, "\n")
cat("Mean: ", mean_orig, "\n")
cat("SD:   ", round(sd_orig, 2), "\n")
cat("Var:  ", round(var_orig, 2), "\n\n")

# Transformation 1: Add 5 to each value
hours_plus5 <- hours + 5
mean_plus5 <- mean(hours_plus5)
sd_plus5 <- sd(hours_plus5)
var_plus5 <- var(hours_plus5)

cat("AFTER ADDING 5 (each student studies 5 more hours):\n")
cat("Data:", hours_plus5, "\n")
cat("Mean: ", mean_plus5, "(increased by 5)\n")
cat("SD:   ", round(sd_plus5, 2), "(UNCHANGED!)\n")
cat("Var:  ", round(var_plus5, 2), "(UNCHANGED!)\n\n")

# Transformation 2: Multiply by 2
hours_times2 <- hours * 2
mean_times2 <- mean(hours_times2)
sd_times2 <- sd(hours_times2)
var_times2 <- var(hours_times2)

cat("AFTER MULTIPLYING BY 2 (double study time):\n")
cat("Data:", hours_times2, "\n")
cat("Mean: ", mean_times2, "(multiplied by 2)\n")
cat("SD:   ", round(sd_times2, 2), "(multiplied by 2)\n")
cat("Var:  ", round(var_times2, 2), "(multiplied by 2Â² = 4!)\n\n")

cat("ðŸ“ TRANSFORMATION RULES:\n")
cat("========================\n")
cat("Adding/Subtracting a constant c:\n")
cat("  â€¢ Mean shifts by c\n")
cat("  â€¢ SD stays the same\n")
cat("  â€¢ Variance stays the same\n\n")

cat("Multiplying/Dividing by a constant c:\n")
cat("  â€¢ Mean multiplied by c\n")
cat("  â€¢ SD multiplied by c\n")
cat("  â€¢ Variance multiplied by cÂ²\n")
````

````r
# Visualize transformations
df_transform <- data.frame(
  value = c(hours, hours_plus5, hours_times2),
  transformation = rep(c("Original", "Add 5", "Multiply by 2"), each = 5)
)

ggplot(df_transform, aes(x = transformation, y = value, fill = transformation)) +
  geom_boxplot(alpha = 0.7) +
  geom_jitter(width = 0.1, alpha = 0.5, size = 2) +
  labs(title = "Effect of Transformations on Data",
       x = "Transformation",
       y = "Value") +
  scale_fill_manual(values = c("Original" = "gray", 
                                "Add 5" = "steelblue", 
                                "Multiply by 2" = "coral")) +
  theme_minimal() +
  theme(legend.position = "none")

cat("\nðŸ“Š Notice:\n")
cat("   â€¢ 'Add 5': Shifts center up, spread UNCHANGED\n")
cat("   â€¢ 'Multiply by 2': Stretches BOTH center and spread\n")
````

## Interactive Exploration Tool

Analyze your own data to see measures of variability!

````r
# Function to analyze variability
analyze_variability <- function(data, title = "Dataset") {
  data <- as.numeric(data)
  
  # Calculate all measures
  n <- length(data)
  mean_val <- mean(data)
  median_val <- median(data)
  range_val <- diff(range(data))
  iqr_val <- IQR(data)
  var_val <- var(data)
  sd_val <- sd(data)
  q1 <- quantile(data, 0.25)
  q3 <- quantile(data, 0.75)
  
  # Print summary
  cat("\n", rep("=", 50), "\n", sep = "")
  cat("VARIABILITY ANALYSIS:", title, "\n")
  cat(rep("=", 50), "\n\n", sep = "")
  
  cat("Sample Size: n =", n, "\n\n")
  
  cat("Measures of Center:\n")
  cat("  Mean:  ", round(mean_val, 2), "\n")
  cat("  Median:", round(median_val, 2), "\n\n")
  
  cat("Measures of Spread:\n")
  cat("  Range:    ", round(range_val, 2), "\n")
  cat("  IQR:      ", round(iqr_val, 2), "\n")
  cat("  Variance: ", round(var_val, 2), "\n")
  cat("  Std Dev:  ", round(sd_val, 2), "\n\n")
  
  cat("Five-Number Summary:\n")
  cat("  Min:", min(data), "\n")
  cat("  Q1: ", q1, "\n")
  cat("  Med:", median_val, "\n")
  cat("  Q3: ", q3, "\n")
  cat("  Max:", max(data), "\n\n")
  
  cat("Coefficient of Variation: CV =", round(sd_val/mean_val * 100, 1), "%\n")
  cat(rep("=", 50), "\n\n", sep = "")
  
  # Create visualizations
  df <- data.frame(value = data)
  
  p1 <- ggplot(df, aes(x = value)) +
    geom_histogram(bins = 10, fill = "steelblue", color = "black", alpha = 0.7) +
    geom_vline(xintercept = mean_val, color = "red", linetype = "dashed", linewidth = 1) +
    geom_vline(xintercept = median_val, color = "blue", linetype = "dashed", linewidth = 1) +
    labs(title = paste("Distribution:", title),
         x = "Value",
         y = "Frequency") +
    theme_minimal()
  
  p2 <- ggplot(df, aes(y = value)) +
    geom_boxplot(fill = "lightsteelblue", color = "black", width = 0.5) +
    labs(title = paste("Boxplot:", title),
         y = "Value") +
    theme_minimal() +
    theme(axis.text.x = element_blank(),
          axis.ticks.x = element_blank())
  
  grid.arrange(p1, p2, ncol = 2)
}

cat("âœ… Function loaded! Use: analyze_variability(your_data, 'Your Title')\n\n")
cat("Example:\n")
cat("my_data <- c(10, 15, 20, 25, 30, 35, 40)\n")
cat("analyze_variability(my_data, 'My Dataset')\n")
````

## Summary and Key Takeaways

### Measures of Variability:

**Range:**
- Easy to calculate: max - min
- Sensitive to outliers
- Uses only 2 data points

**IQR (Interquartile Range):**
- Spread of middle 50% of data: Q3 - Q1
- Resistant to outliers
- Useful with skewed distributions

**Variance (sÂ²):**
- Average squared deviation from mean
- Units are squared (harder to interpret)
- Foundation for many statistical tests

**Standard Deviation (s):**
- Square root of variance
- Same units as original data
- Most commonly reported measure of spread
- Interpretation: "typical distance from the mean"

### Key Insights:

- **Two datasets can have the same mean but very different variability!**
- Low variability â†’ Consistent, predictable
- High variability â†’ Inconsistent, unpredictable
- Use **mean Â± 2s** to estimate where most data falls

### Transformation Rules:

- Adding/subtracting constant: spread UNCHANGED
- Multiplying/dividing by constant: spread changes by same factor
