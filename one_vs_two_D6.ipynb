# Comparing Rolling 1 vs. 2 d6

```r
amount = 20
d6 = sample(rep(1:6, each=50),size=amount)
two_d6 = sample(rep(1:6, each=50),size=amount) + sample(rep(1:6, each=50),size=amount)

paste0("Sample Size:", amount)
hist(d6)
hist(two_d6)
```
